```python
## let's implement AND functionality using the neural net with Theano 
## inputs : binary inputs in the matrix form
## weights: make is a shared variable so that both forward and backprop can update
## output : sigmoid

#import the needed modules
from theano import *   # import theano
import theano.tensor as T

import numpy as np # numpy
import matplotlib.pyplot as plt # for plot
#from random import random

#print(random.__file__)

```

```python
#random.random()
np.random.rand(2)
```

```python
# define the neuron with sigmoid activation
x = T.matrix('x')   #inputs in matrix form
#random.random()
#w = theano.shared(np.array[random.random(), random.random()])
w = theano.shared(np.random.rand(2))
b = theano.shared(1.0)
learning_rate = 0.01
```

```python
# define the mathematical expressions.
z = T.dot(x,w)+b
act = 1/(1+T.exp(-z))

#actual output
a_hat = T.vector('a_hat')

# compute the cost, note that this a logistic model
cost = -(a_hat * T.log(act) + (1-a_hat)* T.log(1-act)).sum()

print(a_hat)
print(cost)

# compute the gradient
dw, db = T.grad(cost, [w,b])

# train
train = function(
       inputs = [x, a_hat],
       outputs = [act, cost],
       updates = [
         [w, w - learning_rate*dw], #weights
         [b, b - learning_rate* db] #bias
    ]
)
```

```python
# define the inputs & outputs for AND
ins = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

#define the outputs
outs = [ 0, 0, 0, 1]

```

```python
#Iterate through all inputs and find outputs:
cost = []
for iteration in range(30000):
    pred, cost_iter = train(ins, outs)
    cost.append(cost_iter)

    #Print the outputs:
print 'The outputs of the NN are:'
for i in range(len(ins)):
    print 'The output for x1=%d | x2=%d is %.2f' % (ins[i][0],ins[i][1],pred[i])
    
    
```
