## Sample  #1 - Read the image file

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
print cv2.__version__
```

```python
## show me the image
## use cv2.imread() for image read
##     cv2.imshow() to show
##     cv2.imwrite() to write to disk

## this will pop up the image.

img = cv2.imread('./hello.jpeg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('My Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## instead of poping up new windows / dialogs, use matplotlib.

img = cv2.imread('./hello.jpeg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()

## write the gray image to disk
cv2.imwrite('./hello_gray.jpeg', img)
```

```python
## show the color image

colr_img = cv2.imread('./hello.jpeg', cv2.IMREAD_COLOR)
#plt.imshow(colr_img, cmap = 'color', interpolation = 'bicubic')
plt.imshow(colr_img)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
