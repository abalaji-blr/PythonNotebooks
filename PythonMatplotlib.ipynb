## More than one plot on same axes

```python
import numpy as np
import random

x1 = random.sample(range(0,10), 5)
print(x1)
x2 = random.sample(range(0,20), 5)
print(x2)

y1 = random.sample(range(-5,5), 5)
y2 = random.sample(range(-10,10), 5)

print(y1)
print(y2)

#set the limits
plt.plot(x1,y1, 'g', linewidth=3)
plt.plot(x2,y2, 'r', linewidth=5)

# set the limits
plt.xlim(0,20)
plt.ylim(-10,10)

plt.title("Two line plots")

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

## Two line plots with legends

```python
import numpy as np
import random

x1 = random.sample(range(0,10), 5)
print(x1)
x2 = random.sample(range(0,20), 5)
print(x2)

y1 = random.sample(range(-5,5), 5)
y2 = random.sample(range(-10,10), 5)

print(y1)
print(y2)

#set the limits
plt.plot(x1,y1, 'g', linewidth=3, label = 'Line one')
plt.plot(x2,y2, 'r', linewidth=3, label = 'Line Two')

# set the limits
plt.xlim(0,20)
plt.ylim(-10,10)

plt.title("Two line plots")

plt.xlabel('x axis')
plt.ylabel('y axis')

# draw the legends
plt.legend()

# draw the grid lines as well
plt.grid(True, color='k')

plt.show()
```

# Scatter Plots

```python
import numpy as np
import random

x1 = random.sample(range(0,10), 5)
print(x1)
x2 = random.sample(range(0,20), 5)
print(x2)

y1 = random.sample(range(-5,5), 5)
y2 = random.sample(range(-10,10), 5)

print(y1)
print(y2)

#set the limits
plt.scatter(x1,y1, color='g')
plt.scatter(x2,y2, color='r')

# set the limits
plt.xlim(0,20)
plt.ylim(-10,10)

plt.title("Two line plots")

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

# Bar Charts

```python
import numpy as np
import random

x1 = random.sample(range(0,10), 5)
print(x1)
x2 = random.sample(range(0,20), 5)
print(x2)

y1 = random.sample(range(0,5), 5)
y2 = random.sample(range(0,10), 5)

print(y1)
print(y2)

#set the limits
plt.bar(x1,y1, color='g', align='center')
plt.bar(x2,y2, color='r', align='center')

# set the limits
plt.xlim(0,20)
plt.ylim(0,10)

plt.title("Two line plots")

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

## Basemap

There are several toolkits for matplotlib. [Visit this link](http://matplotlib.org/2.0.0/mpl_toolkits/index.html) for the info.

Here are some of the available toolkits:
  * Basemap (not part of matplotlib, need to download this)
    + [Intro to Basemap](http://matplotlib.org/basemap/users/intro.html)
    + Inorder to represent the curved surface of the earth in 2D, map projections are used.
      There are [many projections - roughly 33 are available](http://matplotlib.org/basemap/users/mapsetup.html)
      **Following are some of the projections: "mill" (Miller Cylindrical), "eqdc" (Equidistant Conic) etc.**
      **Refer to .../site-packages/mpl_toolkits/basemap/__init__.py**
    + About Coordinates:
      + Note that to identify the location in the map, we have to know the **latitude** (south to north) and **longitude** (west to east).
    
  * Cartopy (alternative mapping toolkit)
  * mplot3D toolkit
  * AxesGrid (to display multiple images)
  * seaborn (for statistical charting)

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
m = Basemap(projection='mill')
m.drawcoastlines()

plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

# plot india with lat & long
m = Basemap(projection='mill',
            llcrnrlat = 6,
            llcrnrlon = 70,
            urcrnrlat = 40,
            urcrnrlon = 100,
            resolution ='c')

m.drawcoastlines()
m.drawmapboundary()
m.fillcontinents(color='coral', lake_color='aqua')
m.drawstates(color='b')
m.drawrivers()

plt.title("India")
plt.show()
```
